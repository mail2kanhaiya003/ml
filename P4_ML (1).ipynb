```python
import matplotlib.pyplot as plt
```

# (A)

```python
#(a) line plot
x = np.arange(0,2*np.pi,0.1) # start,stop,step
y = np.sin(x) # sin() function
z = np.cos(x) #cos() function
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 2pi') # X label
plt.ylabel('sin(x) and cos(x)') # Y Label
plt.title('Plot of sin and cos from 0 to 2pi') #adding title
plt.legend(['sin(x)', 'cos(x)']) # legend entries as seperate v

plt.show()
```

# (B)

```python
#(b) bar plot
# creating the dataset
data = {'Machine Learning':55, 'Artificial Intelligence':55, 'DataMining':30,
 'Project':9,'Information security':16}
subjects = list(data.keys())
students = list(data.values())
 
fig = plt.figure(figsize = (10, 5))
# creating the bar plot
plt.bar(subjects, students, color ='pink',
 width = 0.7)
plt.xlabel("Subjects Offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different Subjects in 6 semester")
plt.show()
```

# (C)

```python
#(c) scatter plot
x = np.random.normal(1.0, 5.0, 1000)
y = np.random.normal(10.0, 20.0, 1000)
plt.scatter(x, y)
plt.show()



```

# (D)

```python
#(d) pie plot
plt.pie(students, labels = subjects)
plt.legend(title = "Five subjects",loc='upper left')
plt.show() 
```

# (E)

```python
#(e) area plot
# Create data
x=range(1,6)
y=[1,4,6,8,4]
# Area plot
plt.fill_between(x, y,color='orange')
plt.show()
```

# (F)

```python
#(f) histogram
x = np.random.normal(170, 10, 250)
plt.hist(x,color='grey')
plt.show() 
```

```python
#Generate different subplots from a given plot and color plot data
# Data for plotting
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.cos(2 * np.pi * x)
# Creating 6 subplots and unpacking the output array
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2)
ax1.plot(x, y, color="violet")
ax2.plot(x, y, color="purple")
ax3.plot(x, y, color="green")
#ax4.plot(x, y, color="magenta")
ax5.plot(x, y, color="black")
ax6.plot(x, y, color="yellow")
```
