## Question 12. Implement Linear Regression problem. For example, based on a dataset comprising of existing set of prices and area/size of the houses, predict the estimated price of a given house.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("USA_Housing.csv")
df.columns = ["Income","Age","Rooms","Bedrooms","Population","Price","Address"]
df.head()
```

### Independent variable = Bedrooms
### Dependent Variable = Price

```python
reg = LinearRegression()
reg.fit(df[['Bedrooms']],df[['Price']])
```

```python
%matplotlib inline
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.scatter(df.Bedrooms,df.Price,color="yellow",marker='.')
plt.plot(df.Bedrooms,reg.predict(df[['Bedrooms']]),color="red")
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
newdf = pd.DataFrame({
    'Bedrooms' : [3,5,6,9,7,10,21]
})
newdf
```

```python
newdf['Price'] = reg.predict(newdf[['Bedrooms']])
newdf
```

```python
%matplotlib inline
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.scatter(newdf.Bedrooms,newdf.Price,color="black",marker='.')
plt.plot(newdf.Bedrooms,reg.predict(newdf[['Bedrooms']]),color="red")
```
